QUESTION THREE: Searching and Sorting Algorithms (6 Marks)

Given the list:

arr = [21, 10, 45, 32, 67, 5, 18]

a) Write a Python function to implement linear search to find the value 32. (2 Marks)

b) Write a Python function to implement binary search to find the value 45.

(Assume the list is sorted before searching.) (2 Marks)

c) Implement bubble sort to sort the list in ascending order. (2 Marks)

```python
# a) Linear Search to find 32
def linear_search(arr, target):
    "Iterates through the list to find the target value."
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

# b) Binary Search to find 45 (Assumes sorted list)
def binary_search(arr, target):
    "Repeatedly divides the search interval in half."
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# c) Bubble Sort implementation
def bubble_sort(arr):
    "Sorts the list by repeatedly swapping adjacent elements."
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

# Test data for search and sorting algorithms
arr = [21, 10, 45, 32, 67, 5, 18]

print(f"Linear Search (32) = Found at index {linear_search(arr, 32)}")

sorted_arr = sorted(arr) # Binary search requires sorting
print(f"Binary Search (45) on {sorted_arr} = Found at index {binary_search(sorted_arr, 45)}")

print(f"Bubble Sort Result = {bubble_sort(arr[:])}")
```

QUESTION FOUR: Linked Lists Implementation (6 Marks)

a) Define a Node class in Python for a singly linked list. (2 Marks)

b) Write a LinkedList class with the following operations: (4 Marks)

i. Insert a node at the beginning
ii. Insert a node at the end
iii. Display all elements in the linked list

```python
# Node class definition
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None    

#b) Linked List class with operations
class LinkedList:
    def __init__(self):
        self.head = None  # Initialize an empty list

    # i. Insert a node at the beginning
    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        print(f"Inserted {data} at the beginning.")

    # ii. Insert a node at the end
    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        
        current = self.head
        while current.next: # Traverse to the last node
            current = current.next
        current.next = new_node
        print(f"Inserted {data} at the end.")

    # iii. Display all elements
    def display(self):
        if not self.head:
            print("The list is empty.")
            return
        
        current = self.head
        print("Linked List Elements:", end=" ")
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

# Execution to demonstrate the operations
my_list = LinkedList()
my_list.insert_at_beginning(10)
my_list.insert_at_end(20)
my_list.insert_at_beginning(5)
my_list.display()
```

QUESTION FIVE: Application of Data Structures (Real-World Problem) (6 Marks)

A university wants to develop a system that manages students waiting to be served at the finance office.

a) Identify the most appropriate data structure to represent this situation and justify your answer. (2 Marks)

b) Write a Python program that simulates this process using the chosen data structure with the following operations: (4 Marks)

i. Add a student to the queue
ii. Serve (remove) the first student in the queue
iii. Display the current queue

a) Queue:It follows the FIFO principle. This mirrors a real-world waiting line where the student who arrives first is the first to be served, ensuring a fair management system.

```python
from collections import deque

class FinanceSystem:
    def __init__(self):
        # Using deque for efficient O(1) additions and removals
        self.queue = deque()

    # i. Add student to queue
    def add_student(self, name):
        self.queue.append(name)
        print(f"Student {name} joined the queue.")

    # ii. Serve (remove) the first student
    def serve_student(self):
        if self.queue:
            served = self.queue.popleft()
            print(f"Serving student: {served}")
        else:
            print("No students in the queue.")

    # iii. Display current queue
    def display_queue(self):
        print(f"Current Queue: {list(self.queue)}")

# Example Usage:
finance = FinanceSystem()
finance.add_student("Enock")
finance.add_student("Kiplangat")
finance.add_student("Rono")
finance.display_queue()
finance.serve_student()
finance.display_queue()
finance.serve_student()
finance.display_queue()
finance.serve_student()
finance.display_queue()
finance.serve_student()
finance.display_queue()
```
